````python
from sqlalchemy import create_engine, MetaData, Table, text, select, func

import csv
import pandas
import sqlite3
from sqlite3 import Error

'''2 Создайте базу данных СУБД SQLite в файле BikeShare.sqlite'''
DataBase = sqlite3.connect('Bikeshare.sqlite')

'''3 Создайте таблицу BikeShare со следующей структурой'''

cur = DataBase.cursor()

create_bikeShare = """

CREATE TABLE BikeShare (
    ride_id text PRIMARY KEY,
    rideable_type text,
    started_at text,
    ended_at text,
    start_station_name text,
    start_station_id integer,
    end_station_name text,
    end_station_id integer,
    start_lat real,
    start_lng real,
    end_lat real,
    end_lng real,
    member_casual text,
    duration integer);
"""

cur.execute(create_bikeShare)

'''4 Наполните БД BikeShare данными из CSV источника.'''

with open('BikeShare_19_20.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)


for i in range(len(data)):
    data[i] = tuple(data[i])


insert = """
INSERT INTO BikeShare (
ride_id,
rideable_type,
started_at,
ended_at,
start_station_name,
start_station_id,
end_station_name,
end_station_id,
start_lat,
start_lng,
end_lat,
end_lng,
member_casual,
duration
)
VALUES(?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
"""
for i in range(1, len(data)):
    DataBase.execute(insert, data[i])
    DataBase.commit()
````

````python
engine = create_engine('sqlite+pysqlite:///Bikeshare.sqlite')
metadata = MetaData()
metadata.reflect(bind=engine)
Bike_Share =Table("BikeShare", metadata, autoload=True, autoload_with="engine")
conn=engine.connect() # Соединение
````

````python
'''5.1. Покажите длительность с накопительным итогом по поездкам,
начавшимся в интервале с 10:00 до 11:30, в порядке возрастания времени начала поездки.'''

s = """
SELECT sum(duration) OVER(order by time(started_at))
FROM BikeShare
WHERE time(started_at) BETWEEN time('10:00:00') AND time('11:30:00')
ORDER BY time(started_at);
"""
with engine.connect() as conn:
    result = conn.execute(s).fetchall()
    print(len(result))
    for row in result:
        print(row)
````

````python
#5.2. Ранжируйте пары станций отправления и назначения по количеству поездок.
# Паре с большим количеством поездок присвойте меньший номер ранга.

s = """
SELECT start_station_name, end_station_name, RANK() OVER(ORDER BY COUNT(ride_id) DESC), COUNT(ride_id)
FROM BikeShare
GROUP BY start_station_name, end_station_name;
"""
with engine.connect() as conn:
    result = conn.execute(s).fetchall()
    print(len(result))
    for row in result:
        print(row)
````

````python
'''5.3 Покажите, сколько в проценках занимает время каждой поездки между начальной и конечной
станциями от суммарного времени, затраченного на все поездки между станциями данной пары'''

s = """
SELECT A.*, ROUND(CAST(A.duration AS REAL)/A.sum_duration * 100, 2)
FROM (
    SELECT BikeShare.start_station_name, BikeShare.end_station_name, BikeShare.duration, sum(BikeShare.duration) OVER (PARTITION BY BikeShare.start_station_id, BikeShare.end_station_id) AS sum_duration
    FROM BikeShare) AS A
ORDER BY A.start_station_name, A.end_station_name;
"""


with engine.connect() as conn:
    result = conn.execute(s).fetchall()
    print(len(result))
    for row in result:
        print(row)
````

````python
'''5.4 Покажите 4 самых долгих поездки с каждой станции отправления
упорядоченные по станции
отправления и по убыванию длительности поездки для данной станции'''

s = """
SELECT A.duration, A.ride_id, A.start_station_name, A.F
FROM (
    SELECT duration, ride_id, start_station_name, RANK () OVER (PARTITION BY start_station_id ORDER BY duration DESC) AS F
    FROM BikeShare order by start_station_name, duration desc) AS A
WHERE A.F<5 order by A.start_station_name, A.duration desc
"""


with engine.connect() as conn:
    result = conn.execute(s).fetchall()
    print(len(result))
    for row in result:
        print(row)

````

````python
'''5.5. Разбейте поездки на квантили по длительности, упорядочив по возрастанию длительности.
Распечатайте следующие поля результата: номер квантиля, идентификатор поездки,
станция назначения, длительность.'''

s = """
SELECT NTILE(4) OVER (ORDER BY duration), ride_id, duration
    FROM BikeShare
"""

with engine.connect() as conn:
    result = conn.execute(s).fetchall()
    print(len(result))
    for row in result:
        print(row)
````
